```python
import requests
import matplotlib.pyplot as plt
import pandas as pd
url = "https://uk.wikipedia.org/wiki/Населення_України"
#Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"
response = requests.get(url)
response.encoding = 'utf-8'  
tables = pd.read_html(response.text, decimal=',', thousands=None)
df = tables[12]
df
```

```python
#Вивести перші рядки таблиці за допомогою методу head
df.head()
```

```python
#Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
df.shape
```

```python
#Замініть у таблиці значення "—" на значення NaN
df = df.replace('—', 'NaN' )
df
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes
df.dtypes
```

```python
#Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
new_df =df.copy()
for colone in df.columns:
    if colone != 'Регіон':
        new_df[colone] = pd.to_numeric(df[colone], errors='coerce')
new_df.dtypes
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
new_df.isnull().sum()
```

```python
#Видаліть з таблиці дані по всій країні, останній рядок таблиці
new_df.drop(27)
```

```python
#Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
avarage =new_df.copy()
for colone in new_df.columns:
    if colone != 'Регіон':
        avarage[colone] = new_df[colone].fillna(new_df[colone].mean().round(1))
avarage
```

```python
#Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
a= avarage.loc[df['Регіон'] == 'Україна', '2019'].values[0]
avarage[avarage['2019'] > a]['Регіон']
```

```python
#У якому регіоні була найвища народжуваність у 2014 році?
avarage[avarage['2014'].max() == avarage['2014']]
```

```python
#Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
plt.bar(avarage['Регіон'], avarage['2019'], color = ["b", "r", "y", "g", "c", "m", "k", "pink", "orange", "purple", "brown", "pink", "lime", "navy", "teal", "aqua", "olive", "maroon", "silver", "gold", "violet", "indigo", "coral", "turquoise", "salmon", "plum", "orchid"])
plt.xlabel('Регіон',fontsize="small", color="midnightblue")
plt.ylabel("Кількість", fontsize="small", color="midnightblue")
plt.title("Коефіцієнт народжуваності в регіонах України (1950—2019)", fontsize=15)
plt.show
```

```python
kir= avarage[avarage['Регіон'] == 'Кіровоградська'].T[10]
kir=kir.drop('Регіон')
plt.plot(kir, color="black")
plt.xlabel('Рік',fontsize='medium', color="orange")
plt.ylabel("Кількість", fontsize="medium", color="red")
plt.title("Коефіцієнт народжуваності в Кіровоградській області", fontsize=15)
plt.show()

```
